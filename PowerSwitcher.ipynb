```python
import requests
import json
import time
import xml.etree.ElementTree as ET

URL = "http://192.168.0.100/restapi/relay/set_outlet_transient_states/"
AUTH = ("admin", "1234")

HEADERS = {
    "X-CSRF": "x",
    "Content-Type": "application/json",
}


def _str_to_bool(s: str) -> bool:
    return s.strip().lower() in ("1", "true", "yes", "on")


def run_sequence_from_xml(xml_path: str) -> None:
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # Initial delay (preserved from your original script)
    time.sleep(5)

    for step in root.findall("Step"):
        switch_str = step.get("switch")
        position_str = step.get("position")
        delay_str = step.get("delay", "0")

        if switch_str is None or position_str is None:
            continue

        try:
            channel = int(switch_str)
        except ValueError:
            continue

        state = _str_to_bool(position_str)

        try:
            delay_ms = int(delay_str)
        except ValueError:
            delay_ms = 0

        payload = [[[channel, state]]]

        try:
            requests.post(
                URL,
                auth=requests.auth.HTTPDigestAuth(*AUTH),
                headers=HEADERS,
                data=json.dumps(payload),
                timeout=5,
            )
        except requests.RequestException:
            pass

        if delay_ms > 0:
            time.sleep(delay_ms / 1000.0)
```

```python
import requests
import time
import xml.etree.ElementTree as ET


def _parse_bool(value: str) -> bool:
    value = (value or "").strip().lower()
    if value in ("true", "1", "yes", "on"):
        return True
    if value in ("false", "0", "no", "off"):
        return False
    raise ValueError(f"Invalid boolean value: {value}")


def run_sequence_xml(
    xml_path: str,
    url: str = "http://192.168.0.100/restapi/relay/set_outlet_transient_states/",
    auth: tuple[str, str] = ("admin", "1234"),
    timeout: int = 5,
    switch_offset: int = -1,   # XML is 1-based, device is 0-based
    dry_run: bool = False,
) -> None:

    headers = {
        "X-CSRF": "x",
        "Content-Type": "application/json",
    }

    root = ET.parse(xml_path).getroot()
    if root.tag != "Sequence":
        raise ValueError(f"Expected <Sequence>, got <{root.tag}>")

    batch = []  # list of [index, bool]

    def flush(delay_ms: int) -> None:
        nonlocal batch
        if not batch:
            return

        # CRITICAL: correct payload shape
        # [
        #   [[idx, bool]],
        #   [[idx, bool]],
        #   ...
        # ]
        payload = [[[idx, state]] for idx, state in batch]

        if dry_run:
            print(f"POST payload: {payload} (then sleep {delay_ms} ms)")
        else:
            resp = requests.post(
                url,
                auth=requests.auth.HTTPDigestAuth(*auth),
                headers=headers,
                json=payload,
                timeout=timeout,
            )
            print(f"Status: {resp.status_code}, Response: {resp.text}")

        if delay_ms > 0:
            time.sleep(delay_ms / 1000.0)

        batch = []

    for step in root.findall("Step"):
        switch_raw = step.get("switch")
        pos_raw = step.get("position")
        delay_raw = step.get("delay", "0")

        if switch_raw is None or pos_raw is None:
            raise ValueError(f"Malformed Step: {ET.tostring(step, encoding='unicode')}")

        idx = int(switch_raw) + switch_offset
        state = _parse_bool(pos_raw)
        delay_ms = int(delay_raw)

        batch.append([idx, state])

        # If delay > 0, include this step, then flush and wait
        if delay_ms > 0:
            flush(delay_ms)

    # Final batch (no delay after)
    flush(0)
```

```python
import time
import requests
import xml.etree.ElementTree as ET


def load_cues(xml_path):
    """
    Parse an XML file and return its root Element.

    :param xml_path: Path to the XML file
    :return: xml.etree.ElementTree.Element
    """
    tree = ET.parse(xml_path)
    root = tree.getroot()

    if root.tag != "Cues":
        raise ValueError(f"Expected root <Cues>, got <{root.tag}>")

    return root



def execute_sequence(
    root,
    base_url="http://192.168.0.100",
    username="admin",
    password="1234",
    timeout=5,
):
    """
    Execute a relay sequence defined by a <Cues> XML root Element.
    """
    url = f"{base_url.rstrip('/')}/restapi/relay/set_outlet_transient_states/"
    auth = requests.auth.HTTPDigestAuth(username, password)

    headers = {
        "X-CSRF": "x",
        "Content-Type": "application/json",
    }

    # Collect and sort cues by order
    cues = []
    for cue in root.findall("Cue"):
        order_raw = cue.get("order")
        if order_raw is None:
            raise ValueError("Cue missing required 'order' attribute")

        try:
            order = int(order_raw)
        except ValueError as e:
            raise ValueError(
                f"Invalid Cue order='{order_raw}' (must be integer)"
            ) from e

        cues.append((order, cue))

    cues.sort(key=lambda t: t[0])

    # Execute each cue
    for _, cue in cues:
        steps = []

        for child in list(cue):
            tag = child.tag
            if not tag.startswith("Switch"):
                continue

            num_raw = tag[len("Switch"):]
            if not num_raw.isdigit():
                continue

            switch_1based = int(num_raw)
            index_0based = switch_1based - 1

            text = (child.text or "").strip().lower()
            if text not in ("true", "false"):
                raise ValueError(
                    f"Invalid value for <{tag}>: '{child.text}' (must be true/false)"
                )

            steps.append([index_0based, text == "true"])

        if not steps:
            raise ValueError(
                "Cue produced no Switch settings (no <SwitchN> elements found)"
            )

        payload = [steps]

        resp = requests.post(
            url,
            auth=auth,
            headers=headers,
            json=payload,
            timeout=timeout,
        )
        resp.raise_for_status()

        # Optional per-cue delay (milliseconds)
        delay_raw = cue.get("delay")
        delay_ms = 0

        if delay_raw is not None:
            delay_raw = delay_raw.strip()
            if delay_raw:
                try:
                    delay_ms = int(delay_raw)
                except ValueError as e:
                    raise ValueError(
                        f"Invalid Cue delay='{delay_raw}' "
                        "(must be integer milliseconds)"
                    ) from e

        if delay_ms > 0:
            time.sleep(delay_ms / 1000.0)
```
